# Problem 1: Smart Light Controller

```python
# Class to represent a Smart Light
class SmartLight:

    # Constructor method to initialize light name and default status
    def __init__(self, light_name):
        self.light_name = light_name   # Store light name
        self.status = "OFF"            # Default status is OFF

    # Method to turn the light ON
    def turn_on(self):
        self.status = "ON"

    # Method to turn the light OFF
    def turn_off(self):
        self.status = "OFF"

    # Method to display current status of the light
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Creating object of SmartLight class
light = SmartLight("Bedroom Light")

# Calling method to turn light ON
light.turn_on()

# Displaying current status
light.display_status()
```

# Problem 2: Employee ID Card System

```python
# Class to represent Employee ID Card
class EmployeeIDCard:

    # Constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name              # Store employee name
        self.emp_id = emp_id          # Store employee ID
        self.department = department  # Store department name

    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Creating object of EmployeeIDCard class
employee = EmployeeIDCard("Rahul", "EMP102", "IT")

# Displaying employee ID card
employee.display_id_card()
```

# Problem 3: Mobile Contact Record

```python
# Class to represent a Mobile Contact
class Contact:

    # Constructor to initialize contact details
    def __init__(self, name, phone):
        self.name = name      # Store contact name
        self.phone = phone    # Store phone number

    # Method to display contact details
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Creating object of Contact class
contact = Contact("Anita", "9876543210")

# Displaying contact details
contact.display_contact()
```

# Problem 4: Product Price Tag Generator

```python
# Class to represent a Product
class Product:

    # Constructor to initialize product details
    def __init__(self, product_name, price):
        self.product_name = product_name  # Store product name
        self.price = price                # Store product price

    # Method to display formatted price tag
    def display_price_tag(self):
        print("Product:", self.product_name)
        print(f"Price: ₹{self.price}")


# Creating object of Product class
product = Product("Headphones", 2499)

# Displaying price tag
product.display_price_tag()
```

# Problem 5: Movie Rating Display System

```python
# Class to represent a Movie
class Movie:

    # Constructor to initialize movie details
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name  # Store movie name
        self.rating = rating          # Store movie rating

    # Method to display movie details
    def display_movie_details(self):
        print("Movie:", self.movie_name)
        print(f"Rating: {self.rating} / 5")


# Creating object of Movie class
movie = Movie("Inception", 4.8)

# Displaying movie information
movie.display_movie_details()
```

# Problem 6: Delivery Address Manager

```python
# Class to represent Delivery Details
class Delivery:

    # Constructor to initialize customer details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name  # Store customer name
        self.address = address              # Store delivery address

    # Method to display delivery details
    def display_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Creating object of Delivery class
delivery = Delivery("Suman", "Hyderabad")

# Displaying delivery information
delivery.display_delivery_details()
```
