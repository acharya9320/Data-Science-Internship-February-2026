
#2. Pass / Fail Analyzer


```python
#Given list of marks
marks=[45,78,90,33,60]
pass_count=0
fail_count=0

#Checking each student's mark
for mark in marks:
  if mark >=50:
    pass_count += 1
  else:
    fail_count += 1

#Printing the final result
print("Total Students: ", len(marks))
print("Total Pass Students: ", pass_count)
print("Total Fail Student: ", fail_count)
```

#3. Simple Data Cleaner

```python
#Given List
names=[" Alice " ,"bob" , "CHARLIE"]

#Declaration
cleaned_names=[]

#Cleaning Process
for name in names:
  cleaned_name=name.strip().lower() #remove extra space
  cleaned_names.append(cleaned_name)
  print("\nCleaned Name: ", cleaned_names)
```

#4. Message Length Analyzer

```python
#Given message list
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
  length=len(message)
  print("Message :", message)
  print("length: ", length)
  if length > 10:
    print("Longer than 10 characters") #flag message
  else:
    print("Shorter message")
  print("=======================================================")
```
