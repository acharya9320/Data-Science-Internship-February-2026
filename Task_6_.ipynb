# Problem Statement 1:Smart Parking Lot Management System

```python
# Function to manage smart parking system
def smart_parking_system(capacity, vehicle_logs):

    # This variable will store current number of parked vehicles
    current_vehicles = 0

    # This variable will store the highest number of vehicles parked at one time
    peak_usage = 0

    # Loop through each log in the list
    for log in vehicle_logs:

        # If vehicle enters, increase count
        if log == "IN":
            current_vehicles += 1

        # If vehicle exits, decrease count
        elif log == "OUT":
            # Make sure vehicles do not go below zero
            if current_vehicles > 0:
                current_vehicles -= 1

        # Check if current vehicles is greater than previous peak
        # If yes, update peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    # Print final number of parked vehicles
    print("Currently Parked Vehicles:", current_vehicles)

    # Print highest parking usage
    print("Peak Parking Usage:", peak_usage)

    # Check parking status
    if current_vehicles > capacity:
        print("Parking Status: FULL - Capacity Exceeded!")
    else:
        print("Parking Status: Available")


# ---------------------------
# Sample Input
# ---------------------------

capacity = 50

vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Call the function
smart_parking_system(capacity, vehicle_logs)
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    # 1. Base time: 3 minutes per kilometer
    base_time_per_km = 3
    estimated_time = distance * base_time_per_km

    # 2. Add fixed preparation time (Restaurant side)
    estimated_time += 15

    # 3. Traffic Adjustments
    traffic_delays = {
        "Low": 0,
        "Medium": 5,
        "High": 12
    }
    estimated_time += traffic_delays.get(traffic, 0)

    # 4. Weather Adjustments
    weather_delays = {
        "Clear": 0,
        "Rainy": 4,
        "Stormy": 15
    }
    estimated_time += weather_delays.get(weather, 0)

    return f"Estimated Delivery Time: {estimated_time} minutes"

# Testing with Sample Input
dist = 8
traff = "High"
weath = "Rainy"

print(estimate_delivery_time(dist, traff, weath))
```

## Problem 2 : second method

```python
def estimate_delivery_time(distance, traffic, weather):
    base_speed = 30
    base_time = (distance / base_speed) * 60

    # Traffic delay
    traffic_delay = {"low": 5, "medium": 10, "high": 20}

    # Weather delay
    weather_delay = {"clear": 0, "rainy": 15, "stormy": 25}

    buffer_time = 4  # Safety buffer

    final_time = base_time \
                 + traffic_delay.get(traffic.lower(), 0) \
                 + weather_delay.get(weather.lower(), 0) \
                 + buffer_time

    return round(final_time)


print("Estimated Delivery Time:",
      estimate_delivery_time(8, "High", "Rainy"),
      "minutes")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats):

    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Filling Fast"
    else:
        status = "Seats Available"

    # Suggest additional show
    if occupancy >= 90:
        suggestion = "Consider Opening Additional Show"
    else:
        suggestion = "No Extra Show Required"

    return round(occupancy), status, suggestion


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

result = analyze_seat_occupancy(total_seats, booked_seats)

print("Occupancy:", str(result[0]) + "%")
print("Show Status:", result[1])
print("Suggestion:", result[2])
```

# Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):

    # Calculate average CPU usage
    total = 0
    for value in cpu_readings:
        total += value

    average_cpu = total / len(cpu_readings)

    # Determine server status
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu < 80:
        status = "Warning"
    else:
        status = "Critical"

    return round(average_cpu), status


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

result = classify_server_load(cpu_readings)

print("Average CPU Load:", str(result[0]) + "%")
print("Server Status:", result[1])
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def monitor_resource_usage(resource_usage):

    # Define maximum allowed usage hours
    threshold = 8

    overused_resources = []

    # Check each resource
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)

    # Energy alert condition
    energy_alert = "Yes" if overused_resources else "No"

    return overused_resources, energy_alert


# Sample Input
resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

result = monitor_resource_usage(resource_data)

print("Overused Resources:",
      ", ".join(result[0]) if result[0] else "None")
print("Energy Alert:", result[1])
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_event_registration(capacity, registrations):

    confirmed = 0
    waitlisted = 0

    # Loop to simulate registration processing
    for i in range(1, registrations + 1):
        if i <= capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Registration status
    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"

    return confirmed, waitlisted, status


# Sample Input
capacity = 100
registrations = 105

result = manage_event_registration(capacity, registrations)

print("Confirmed Registrations:", result[0])
print("Waitlisted Users:", result[1])
print("Registration Status:", result[2])
```
